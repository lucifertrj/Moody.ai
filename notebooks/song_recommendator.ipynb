```python
import numpy as np
```

```python
numer_cols = ['length','popularity',"instrumentalness","liveness","valence","loudness","speechiness",'tempo','danceability']
```

```python
plt.figure(figsize=(18,10))
for i,col in enumerate(numer_cols,start=1):
    plt.subplot(3,3,i)
    plt.scatter(range(data.shape[0]), np.sort(data[col].values))
    plt.title(col)
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()
```

```python
data[numer_cols] = minmax.fit_transform(data[numer_cols])
```

```python
data[numer_cols]
```

```python
plt.figure(figsize=(18,10))
for i,col in enumerate(numer_cols,start=1):
    plt.subplot(3,3,i)
    plt.scatter(range(data.shape[0]), np.sort(data[col].values))
    plt.title(col)
```

```python
data.describe()
```

```python
plt.figure(figsize=(18,10))
for i,col in enumerate(numer_cols,start=1):
    plt.subplot(3,3,i)
    sns.distplot(np.sort(data[col].values))
    plt.title(col)
```

```python
for col in numer_cols:
    print(f"{col}:{data[col].skew()}")
```
